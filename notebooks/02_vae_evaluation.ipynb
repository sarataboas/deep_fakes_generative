```python
import pandas as pd
results_dir = "../results/"
```

```python
results_vae_baseline = pd.read_csv(results_dir + "vae_latent256_beta2.csv")

# create me a plot with the values in results
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.plot(results_vae_baseline["epoch"], results_vae_baseline["train_loss"],
            label="Train Loss", color="lightblue")
plt.plot(results_vae_baseline["epoch"], results_vae_baseline["val_loss"],
            label="Validation Loss", color="lightcoral")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("VAE Baseline Loss Curves")
plt.legend()
plt.grid()
plt.show()
```
