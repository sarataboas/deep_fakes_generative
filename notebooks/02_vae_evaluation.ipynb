### **VAE Training Plots**

This notebook contains the Variational Autoencoder training evolution of the several experiments done during this project.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### **Experiments**:

1. Baseline VAE: ***01_baseline_vae.json***  

2. Baseline VAE with KL Annealing: ***02_baseline_vae_kl_annealing.json***  

3. Baseline VAE with Increased Latent Capacity: ***03_baseline_vae_latent_capacity.json***  

4. New architecture (5 layer) VAE: ***04_5layer_vae_big_imgsize.json***  

5. Face Cropped Data Baseline VAE: ***05_facecrop_data_baseline_vae_kl_annealing.json***  

6. Baseline VAE with Perceptual Loss: ***06_baseline_vae_perceptual_loss.json***  

7. Baseline VAE with Multi-Layer Perceptual Loss: ***07_baseline_vae_multilayer_perceptual_loss.json***  

8. Baseline VAE with Multi-Layer Perceptual Loss and Upsampling+Convolution Layers: ***08_baseline_vae_perceptual_loss_upsampling+convolution.json***  


To visualize a specific configuration training evolution, please change the `selected_config` name for one of the listed ones.

```python
selected_config = "01_baseline_vae.csv"
```

```python
results_dir = "../results/"
results_vae_baseline = pd.read_csv(results_dir + selected_config)
```

#### **1. Training and Validation Loss Visualization**

```python
sns.set_theme(style="whitegrid", context="talk")

# Create figure
plt.figure(figsize=(12, 6))

# Plot curves
plt.plot(
    results_vae_baseline["epoch"],
    results_vae_baseline["train_loss"],
    label="Train Loss",
    linewidth=2.5,
    color="#6D97DB"
)

plt.plot(
    results_vae_baseline["epoch"],
    results_vae_baseline["val_loss"],
    label="Validation Loss",
    linewidth=2.5,
    color="#D03B5E"
)

# Labels and title
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("Loss", fontsize=14)
plt.title("New Data (Face Cropped) Training vs Validation Loss", fontsize=18, weight="bold")

# Legend
plt.legend(frameon=True, fancybox=True, shadow=True)

# Improve ticks
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Remove top/right borders
sns.despine()

# Tight layout
plt.tight_layout()

# Show plot
plt.show()
```

#### **2. Training and Validation Reconstruction Loss Visualization**

```python
results_vae_baseline = pd.read_csv(results_dir + "face_crop_vae.csv")

import matplotlib.pyplot as plt
import seaborn as sns

# Style
sns.set_theme(style="whitegrid", context="talk")

# Create figure
plt.figure(figsize=(12, 6))

# Plot curves
plt.plot(
    results_vae_baseline["epoch"],
    results_vae_baseline["train_reconstruction_loss"],
    label="Train Reconstruction Loss",
    linewidth=2.5,
    color="#6D97DB"
)

plt.plot(
    results_vae_baseline["epoch"],
    results_vae_baseline["val_reconstruction_loss"],
    label="Validation Reconstruction Loss",
    linewidth=2.5,
    color="#D03B5E"
)

# Labels and title
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("Loss", fontsize=14)
plt.title("New Data (Face Cropped) Training vs Validation Reconstruction Loss", fontsize=18, weight="bold")

# Legend
plt.legend(frameon=True, fancybox=True, shadow=True)

# Improve ticks
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Remove top/right borders
sns.despine()

# Tight layout
plt.tight_layout()

# Show plot
plt.show()
```
