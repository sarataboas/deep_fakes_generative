```python
import pandas as pd
results_dir = "../results/"
```

```python
results_vae_baseline = pd.read_csv(results_dir + "face_crop_vae.csv")

import matplotlib.pyplot as plt
import seaborn as sns

# Style
sns.set_theme(style="whitegrid", context="talk")

# Create figure
plt.figure(figsize=(12, 6))

# Plot curves
plt.plot(
    results_vae_baseline["epoch"],
    results_vae_baseline["train_loss"],
    label="Train Loss",
    linewidth=2.5,
    color="#6D97DB"
)

plt.plot(
    results_vae_baseline["epoch"],
    results_vae_baseline["val_loss"],
    label="Validation Loss",
    linewidth=2.5,
    color="#D03B5E"
)

# Labels and title
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("Loss", fontsize=14)
plt.title("New Data (Face Cropped) Training vs Validation Loss", fontsize=18, weight="bold")

# Legend
plt.legend(frameon=True, fancybox=True, shadow=True)

# Improve ticks
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Remove top/right borders
sns.despine()

# Tight layout
plt.tight_layout()

# Show plot
plt.show()
```
